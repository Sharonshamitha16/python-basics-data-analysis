```python
def set_operation(sent1,sent2):
    ''' input:sent1,sent2-two sentences taken as inputs
         output: return the sum of the length of unique words.'''
    collectsent = sent1 + " " + sent2 
    collectsentsplit= collectsent.split()
    collectsentappend=[]
    for i in range  (0,len(collectsentsplit)-1):
         if collectsentsplit[i] not in collectsentappend:
             collectsentappend.append(i)
             w= len(collectsentappend)
    total=0
    for j in w:
        total+=len(j)
    return total
```

```python
class Human:
    Population =0
    data=[]
    def __init__(self,name,age,alive=True):
        self.name= name
        self.age = age
        self.alive=alive
        Human.Population+=1
        Human.data.append(self.name)

    def greet(self):
        print("Hey my name is" ,self.name)
    
    def dead(self):
        if self.alive:
            print(self.name, "is dead now")
            Human.Population-=1
            # Human.data.remove(self.name)
            self.alive=False
        else:
            print(self.name, "is already dead")
    def child(self, number):
        Human.Population+=number
        Human.data.append(self.name + "'s child")

sharon = Human("sharon",24)
sharon.name
Human.Population
sam = Human("sam", 30)
sam.name
Human.Population
Human.data
sharon.dead()


```
